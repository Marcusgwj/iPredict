```python
import pandas as pd
import yfinance as yf
import datetime as dt
import time
import requests
import io
```

```python
# Ask user to input?
start = dt.datetime(2022,2,1)
end = dt.datetime(2022,3,1)
```

```python
# Can the website work if use csv file like this?
sgx = pd.read_csv("SGXData_mainboard_active.csv")
sp500 = pd.read_csv("constituents.csv")
```

```python
# Getting the Trading Code only and turn it into a list of string

Symbols = sgx['Trading Code'].tolist()

# for sgx, need to add .SI in the end
def add_SI(x):
  new_lst = []
  for ele in x:
    ele = ele + '.SI'
    new_lst.append(ele)
  return new_lst

Symbols = add_SI(Symbols)
Symbols.insert(0, "^STI")

# extend the list with SP500
Symbols.extend(sp500['Symbol'])
Symbols
```

```python
# GET DATA FROM YAHOO FINANCE!
# Not sure if we want to use this code. Only use this if you want to get the stock prices for ALL of the stock codes
# Maybe can use if we update our stock price database daily, and that we retrieve the information from our database when the user calls it?
# Or is it better to retrieve information from yahoo finance directly every time the user calls it.
# Runtime: 13m 51s

# create empty dataframe
stock_final = pd.DataFrame()
# iterate over each symbol
for i in Symbols:  
    
    # print the symbol which is being downloaded
    print( str(Symbols.index(i)) + str(' : ') + i, sep=',', end=',', flush=True)  
    
    try:
        # download the stock price 
        stock = []
        stock = yf.download(i,start=start, end=end, interval='1d', progress=False)
        
        # append the individual stock prices 
        if len(stock) == 0:
            None
        else:
            stock['Name']=i
            stock_final = stock_final.append(stock,sort=False)
    except Exception:
        None
```

```python
# GET THE CLOSE STOCK PRICE FOR EACH DATE ONLY

df = stock_final
df.reset_index(inplace=True)
df = df.filter(['Name', 'Date', 'Close'])
df.head()
```

```python
# GET STOCK PRICE DATA FROM YAHOO FINANCE DIRECTLY EVERY TIME USER CALLS FOR IT

def stock_price_history(stock_code, start, end, interval):
  try:
    # download the stock price 
    result = []
    result = yf.download(stock_code, start=start, end=end, interval='1d', progress=False)

    # if there is no result in this period, return None
    if len(result) == 0:
      None

  except Exception:
    None

  # Just take the dates and close price for prediction  
  return result.filter(['Date', 'Close'])

stock_price_history('1B0.SI', dt.datetime(2022,2,1), dt.datetime(2022,3,1), '1d')
```
